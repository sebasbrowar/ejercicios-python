## Creando un dataframe

Pandas es un marco de desarrollo para la manipulación de datos, creado originalmente para ofrecer en python las facilidades de manejo de tablas de datos que tiene en forma nativa el lenguaje `R`.

Pandas se basa en dos clases: `Serie` y `DataFrame`, ambas heredadas de la clase `ndarray`de numpy. Un objeto de la clase `Serie`(que llamaremos serie) es un arreglo de datos de un solo tipo, los cuales se encuentran indexados. Un objeto tipo `DataFrame`es una colección de series, en la que se comparte el índice (o renglón), pero cada serie (o columna) tiene su propio tipo. En la figura se muestra como es un DataFrame

![](https://pandas.pydata.org/docs/_images/01_table_dataframe.svg)

Vamos a crear un `Dataframe`:

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"]
    }
)
```

## Seleccionando partes de un dataframe

### Seleccionando columnas

Seleccionar columnas es muy fácil, solo hay que tener presente que si se selecciona una sola columna, lo que se obtiene es una serie, mientras que si se selecciona un subconjunto de columnas, lo que se obtiene es otro dataframe. Quedarse con un subconjunto de columnas se conoce tambien como seleccionar.

### Seleccionando renglones

Los renglones tienen mas detallitos a tomar en cuenta que las columnas.
Este proceso se conoce en general como filtrado, y lo que se busca es seleccionar solo los
renglones que cumplan ciertos criterios. Veamos.

Vamos empezando por buscar un dataframe de la información de los pasageros con 35 años o mas:

```python
df_35mas = df_titanic[df_titanic['Age'] > 35]   #equivalente df_titanic.Age[df_titanic.Age > 35]

df_35mas.describe()
```

Ahora vamos a buscar los pasajeros que se viajaron en 1ra o 2da clase:

```python
df_12 = df_titanic[df_titanic.Pclass.isin([1, 2])]

df_12.describe()
```

Se pueden usar combinadores lógicos entre diferentes columnas, pero las condiciones debe estar
clara con el uso de paréntesis y se deben utilizar `|` para la disyunción y `&` para la conjunción.

Vamos a ver los pasajeros de mñas de 35 años y que viajen en 1ra o 2da:

```python
df_titanic[(df_titanic.Age > 35) & (df_titanic.Pclass.isin([1, 2]))].describe()
```

Por últomo vamos a ver los pasajeron que no registraron su edad:

```python
df_sin_edad = df_titanic[df_titanic.Age.isna()]

df_sin_edad
```

### Seleccionando renglones y columnas

![](https://pandas.pydata.org/docs/_images/03_subset_columns_rows.svg)

Este paso es un poco extraño, ya que no se pueden seleccionar renglones y columnas directamente, sino que hay que usar los métodos `.loc`y `.iloc`. Vamos a ejemplificarlos.

Supongamos que queremos los nombres de todos los mayores de 35 años, como serie y como dataframe. La manera de seleccionar y filtrar es la siguiente:

```python
como_serie = df_titanic.loc[df_titanic.Age > 35, 'Name']  # Como serie de tiempo

como_df = df_titanic.loc[df_titanic.Age > 35, ['Name']]  # Como dataframe
```

```python
como_serie
```

```python
como_df
```

Si quieres no solo el nombre, pero tambien el género se puede obtener como:

```python
df_nueva = df_titanic.loc[df_titanic.Age > 35, ['Name', 'Sex']]
df_nueva
```

Este tipo de indexación tambien puede servir para modificar valores, por ejemplo:

```python
df_nueva.loc[df_nueva.Sex == 'female', 'Sex'] = 'mujer'
df_nueva
```

## Extrayendo estadísticas

### Por columna

Para encontrar estadísticas podemos seleccionar una columna y aplicarle cualquier oeración de agregación incluida en las operaciones en series. Por ejemplo, para encontrar la edad promedio de los pasajeros del titanic:

```python
df_titanic['Age'].mean()
```

Las operaciones de agregación tambien se pueden aplicar en dataframes, aplicandose en cada serie en forma independiente. Por ejemplo:

```python
df_titanic[['Age', 'Fare']].median()
```

Y si se quiere aplicar una serie de agregaciones diferentes a varias variables, lo mejor es usar el método `.agg`como se muestra a continuación:

```python
df_titanic.agg(
    {
        "Age": ["min", "max", "median"],
        "Fare": ["min", "max", "mean", "std"],
    }
)
```

### Regrupando por variables

Para esto se usa el método `.group` que si bien de inicio parece bastante obvio, luego se le ven algunos detallitos.

Por ejemplo, supongamos que quiero saber la edad pronedio de los pasajeros por género. Hay dos maneras de hacerlo. La primera es seleccionar `Age` y `Sex`, regrupar por `Sex`y sacarle la media a Àge`.

![](https://pandas.pydata.org/docs/_images/06_groupby.svg)

La otra es regrupar por `Sex`, luego seleccionar `Age`, y aplicarle la media.

![](https://pandas.pydata.org/docs/_images/06_groupby_select_detail.svg)

Vamos a ver que sale en cada caso:

```python
df_titanic[['Age', 'Sex']].groupby('Sex').mean()
```

```python
df_titanic.groupby('Sex')[['Age']].mean()
```

Se pueden hacer regrupaciones en mútiples niveles. Por ejemplo si queremos saber la edad promedio, por género y por clase en la que viajaban, se puede regrupar en dos variables:

```python
df_titanic.groupby(["Sex", "Pclass"])[["Age"]].mean()
```

```python
import pandas as pd

# Cargar un DataFrame desde un archivo CSV
df = pd.read_csv('https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv')
```

```python
# Por comodidad y para que se vea mejor, creare el indice y se llamara id

df.rename(columns={'Unnamed: 0': 'id'}, inplace=True) # Cambia el nombre de la colmuna pasada que era como el indice y lo renombra a id, sin crear un nuevo dataframe con el inplace=True
df.set_index('id', inplace=True) # Asignamos el nuevo indice llamado id, sin crear un nuevo dataframe con el inplace=True
```

```python
# ¿Cuantas variables tiene el dataframe?

# Podemos simplemente escribir df para ver las primeras filas, y vemos que tiene 13 variables, sin contar la id.
df
```

```python
# ¿Qué variables tienen valores perdidos?

# Para encontrar los valores perdidos podemos utilizar .isnull y nos devuelve un dataframe con valores booleanos, si hay un null tendra True, si no hay tendra False
valores_perdidos = df.isnull()

# Y con .sum podemos contar los valores perdidos en cada columna, o sea contara los True ya que True se considera como 1 y False como 0
conteo_perdidos = valores_perdidos.sum()

print("Valores perdidos por columna:")
print(conteo_perdidos)
# De esta manera podemos ver que las variables country, designation, price, province, region_1, region_2, taster_name, taster_twitter_handle tienen valores perdidos
```

```python
# ¿Qué variables son numéricas?

# Las variables numericas son points y price, sin contar a unnamed.
# Podemos obtenerlas de esta manera

# Utilizamos .dtypes que nos devuele una serie con los tipos de datos en el dataframe
tipos = df.dtypes

# Ahora hacemos otra pero filtrando los tipo object para que nos queden los numericos
columnas_numericas = tipos[tipos != 'object']

print("Variables numéricas:")
print(columnas_numericas)
```

```python
# ¿Qué variables son cualitativas?

# Podemos hacer los mismo que con los numericos pero al reves

# Filtramos los tipo numericos para que nos queden los object
columnas_object = tipos[tipos != 'int64'][tipos != 'float64']

print("Variables cualitativas:")
print(columnas_object)
```

```python
# Hacer un dataframe con únicamente vinos europeos.

# Primero voy a hacer un conteo de los paises para ver cuantos y cuales son
paises = df['country']

# Utilizamos .value_counts() ya que nos crea una serie con el numero de frecuencias en orden de mayor a menor de lo que se le asigne, en este caso la serie con solo los paises
conteo_paises = paises.value_counts()

print(conteo_paises)
```

```python
# Creamos una lista con solo los paises europeos
europa = ['France', 'Italy', 'Spain', 'Portugal', 'Austria', 'Germany', 'Greece',
      'Bulgaria', 'Romania', 'Hungary', 'Croatia', 'Turkey', 'Slovenia', 'Georgia',
      'England', 'Moldova', 'Serbia', 'Macedonia', 'Czech Republic', 'Ukraine',
      'Switzerland', 'Luxembourg', 'Bosnia and Herzegovina', 'Slovakia']

# Creamos el dataframe con solo los paises europeos utilizando .isin() que hace que la columna country solo tenga lo que se encuentra en la lista europa
df_eu = df[df['country'].isin(europa)]

df_eu
```

```python
# ¿Cuál es el menor, el mayor y el precio promedio de la botella por país?
# Esto lo podemos averiguar con .describe() que nos brinda la informacion que buscamos y mas
# Observamos que el menor precio es 4, el mayor es 2500 y el precio promedio es 35.23

df.describe()
```

```python
# ¿De que país es la botella de menor precio?

# Podemos utilizar .idxmin que nos devuele el indice donde se encuentra el valor minimo de la clomuna asignada, en este caso la columna es price
indice_min_precio = df['price'].idxmin()

# Ahora con .iloc accedemos a esa fila con el indice que obtuvimos antes
vino_menor_precio = df.iloc[indice_min_precio]

# Observamos que la botella con menor precio es de España, y tiene un precio de 4 el cual ya vimos que sí es el menor
print(vino_menor_precio)
```

```python
# ¿Cuantos vinos hay con *aroma a fresa* entre otras consideraciones snobs que vienen en la descripción?

# Anteriormente vimos que no habia valores perdidos en la columna de description, por lo que todos los vinos tienen una descripcion
# Entonces utilizaremos .count() para contar los valores de la columna desciption y nos aprovechamos que no cuenta los valores nulos
conteo_vinos_descr = df['description'].count()
print(conteo_vinos_descr)
```

```python
# ¿Cuantas designaciones diferentes hay? ¿Cuál es la más repetida? ¿Cuantas veces se repite?

# Primero creare una serie con solo las designaciones
designaciones = df['designation']

# Utilizamos .value_counts() ya que nos crea una serie con el numero de frecuencias en orden de mayor a menor de lo que se le asigne, en este caso la serie con las designaciones
conteo_designaciones = designaciones.value_counts()

# Al imprimir esto, podemos deducir que la length es el numero de diferentes designaciones, entonce hay 24,187 designaciones diferentes
# Observamos que la mas repetida es la Reserve y se repite 999 veces
print(conteo_designaciones)
```

```python
# Hacer un dataframe con la variedad, el país y el precio para vinos con un costo menor a los $20 dolares

# Primero creamos un nuevo dataframe con solo las columnas que nos interesan
df_2 = df[['variety', 'country', 'price']]

# Ahora modificamos el dataframe para que los valores en la columna de price sean menores a 20 y esto hace que se filtren
df_2 = df_2[df_2['price'] < 20]

# Tambien por las dudas podemos eliminar las filas que contienen nulos en la columna price con .dropna
df_2 = df_2.dropna(subset=['price'])

# Imprimimos el dataframe para ver que esta ok y con .describe() observamos que el maximo es 19, por lo que el dataframe esta correcto
print(df_2)
df_2.describe()
```

```python
# ¿Cuantos vinos diferentes de la variedad *Pinot Noir* hay por cada país?

# Primero creamos un dataframe con solo country y variety, donde en la variedad solo tendremos Pinot Noir
# De esta manera tendremos un dataframe el cual solo tendra paises con variedad = Pinot Noir
df_pinot = df.loc[df.variety == 'Pinot Noir', ['country', 'variety']]

# Ahora creamos una serie con solo los paises
pinot_paises = df_pinot['country']

# Con .value_counts() contamos la frecuencia de aparicion de cada pais, entonces tendremos una serie con el conteo de variedad de Pinot Noirs por pais
conteo_pinot_paises = pinot_paises.value_counts()

print("Variedad de Pinot Noir por cada pais:")
print(conteo_pinot_paises)
```
